# Extract images from video OpenCV
This program extract the images from video and save them in a single folder in .jpeg format using Python OpenCV library.

```python
#import libraires
import cv2
import os
```

```python
camera = cv2.VideoCapture("C:\\Research Work\\vid.mp4") 
  
try:  
    # creating a folder named data 
    if not os.path.exists('Images'): 
        os.makedirs('Images') 
# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of Images')   
# frame counter
currentframe = 0  
while(True):     
    # reading from frame 
    ret,frame = camera.read() 
    if ret: 
        # if video is still left continue creating images 
        name = './Images/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name) 
        # writing the extracted images 
        cv2.imwrite(name, frame) 
        # increasing counter so that it will show how many frames are created 
        currentframe += 1
    else: 
        break 
camera.release() 
cv2.destroyAllWindows() 
```
